```python
def fibonacci(n):
    #for a given n, the output should be the nth fibonacci number
    #e.g. fibonacci(4 should return 3
    if (n==1) or (n==2):
        return 1
        
    if n>2:
        return fibonacci(n-1)+fibonacci(n-2)
    
```
